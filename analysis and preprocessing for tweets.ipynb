#   Text_Cleaning_and_Normalization
#### حذف الروابط والصور والحروف المكررة وعلامات الترقيم
#### وتبديل بعض الحروف كالهمزة و... بأحرف بديل 
#### حذف المنشن والريتويت Mention And Retweet
#### حذف علامات التشكيل



```python
#Text_Cleaning_and_Normalization
def normalize(text):
    text = re.sub("ة", "ه",text)
    text = re.sub("[إأآا]", "ا",text)
    text = re.sub("ى", "ي",text)
    text = re.sub("ئ", "ء",text)
    text=re.sub(r'(.)\1+', r'\1\1',text)#حذف الحروف المكررة مع اللإبقاء على حرفين فقط من أجل مثلا كلمة الله
    return text


def Text_Cleaning_and_Normalization(text):
    URLPic=r'pic.twitter.com/[A-Za-z0-9]+'#حذف الصور 
    URLLink=r'http(s)?://(\s)*[A-Za-z]+\.[A-Za-z]+((/)?([A-Za-z0-9])+(\s)?)+'#حذف الروابط
    punctuations_Symbols = r'[\.؟!;()&,%"''?]'#حذف المحارف غير المرغوب بها كعلامات الترقيم
    arabic_diacritics = r' َ  ُ ِ ّ  ً  ٌ ٍ  ْ '#حذف علامات الشكيل
    EngNames=r'@[A-Za-z0-9]+'#  بالانكليزي mention  وال Retweetحذف ال   
    ArbNames=r'[0-9_ا-ي]+@'#    بالعربيmention  وال Retweetحذف ال 

    
    text=re.sub(URLLink,'',text)
    text=re.sub(URLPic,'',text)
    text=re.sub(punctuations_Symbols,'',text)
    text=re.sub(arabic_diacritics,'',text)
    text=re.sub(EngNames,'',text)
    text=re.sub(ArbNames,'',text)
    text=normalize(text)
    return text

f=open("tweets.txt", "r",encoding='utf-8')
text =f.read()
text=Text_Cleaning_and_Normalization(text)
text=normalize(text)
```

#  Text_Analysis
Tokenizingتحويل كل تويتاية لمجموعة من الكلمات حسب المسافة الفاصلة بين الكلمات 
StopWrodsحذف الكلمات الغير هامة
Steemingارجاع الكلمات للجذر


```python
#Text_Analysis
stop_words = set(stopwords.words('arabic'))
def Text_Analysis(text):
    tokenizer= RegexpTokenizer('\n|\s', gaps=True)
    tokens=tokenizer.tokenize(text)
    tokens=[w for w in tokens if not w in stop_words]

    st = ISRIStemmer()
    stem_tokens=[st.stem(w)for w in tokens]
    stem_tokens
    return stem_tokens
    
stem_tokens=Text_Analysis(text)  
stem_tokens
```

# ثنائيات الكلمات الشائعة

```python
#ثنائيات الكلمات الشائعة 
bigrams = nltk.bigrams(stem_tokens)
s_cfd = nltk.ConditionalFreqDist(bigrams)
PairFdist = nltk.FreqDist(s_cfd)
PairFdist.most_common(5)

```

# ثلاثيات الكلمات الشائعة  

```python
#ثلاثيات الكلمات الشائعة 

trgrams = nltk.trigrams(stem_tokens)
sS_cfd = nltk.ConditionalFreqDist((m[0],(m[1],(m[2])))for m in trgrams)
THirdFdist=nltk.FreqDist(sS_cfd)
THirdFdist.most_common(5)
```

# Collocations

```python
#Collocations

from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()

finder = BigramCollocationFinder.from_words(stem_tokens)
finder2 = TrigramCollocationFinder.from_words(stem_tokens)

scored = finder.score_ngrams(bigram_measures.raw_freq)
scored2 = finder2.score_ngrams(trigram_measures.raw_freq)

sorted(bigram for bigram, score in scored)
#sorted(trigram for trigram, score in scored2)

```

# ِClustering with kmeans 

```python
#ِClustering

from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import collections
def cluster_sentences(sentences, nb_of_clusters=10):
        tfidf_vectorizer = TfidfVectorizer(max_df=0.5,min_df=0.02) 

        tfidf_matrix = tfidf_vectorizer.fit_transform(sentences) 
        kmeans = KMeans(n_clusters=nb_of_clusters)
        kmeans.fit(tfidf_matrix)
        clusters = collections.defaultdict(list)
        for i, label in enumerate(kmeans.labels_):
                clusters[label].append(i)
        return dict(clusters)
    
data=text.split("\n")   
s=cluster_sentences(data,3)
```
